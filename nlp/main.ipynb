```python
from tools import data_parse,identify
from tools import check
def main(input_file, output_file):
    # 按100字分割文本
    text_chunks = data_parse.split_document_into_chunks(input_file)
    
    # 识别多音字和易读错字并添加注音
    annotated_chunks = []
    for chunk in text_chunks:
        print(chunk)
        print('----------------------------')
        annotated_text = identify.identify_issues(chunk)
        print(annotated_text)
        annotated_chunks.append(annotated_text)
    
    # 生成新的Word文档
    data_parse.create_new_word_document(annotated_chunks, output_file)
    # 老师、学生分别审阅文档给出意见
    student = check.Student(output_file)
    student.review_document()

    teacher = check.Teacher(output_file)
    teacher.review_document()

if __name__ == "__main__":
    input_file = "data/副本人机交互选题.docx"
    output_file = "data/output.docx"
    main(input_file, output_file)
```
